```python
from sklearn.preprocessing import StandardScaler,RobustScaler
```

```python
std_scaler = StandardScaler()
rob_scaler = RobustScaler()
```

```python
data['amount_scale'] = rob_scaler.fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
data['time_scale'] = rob_scaler.fit_transform(data['Time'].values.reshape(-1,1))
```

```python
data.head(5)
```

```python
data.drop(['Time','Amount'],axis=1,inplace=True)
```

```python
data.info()
```

```python
amount_scale = data['amount_scale']
time_scale = data['time_scale']
```

```python
data.drop(['amount_scale', 'time_scale'], axis=1, inplace=True)
data.insert(0, 'amount_scale', amount_scale)
data.insert(1, 'time_scale', time_scale)
```

```python
data.head(5)
```

```python
data = data.sample(frac=0.5)
```

```python
data.head(5)
```

```python
fraud = data.loc[data['Class'] == 1]
no_fraud = data.loc[data['Class'] == 0][:492]
```

```python
normal_distributed_data = pd.concat([fraud, no_fraud])
```

```python
sample_data = normal_distributed_data.sample(frac=1, random_state=42)
sample_data.head()
```

```python
X = sample_data.drop('Class' , axis = 1)
y = sample_data['Class']
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC()
```

```python
clf.fit(X_train, y_train)
```

```python
y_train_pred = clf.predict(X_train)
```

```python
print(accuracy_score(y_train, y_train_pred))
```

```python
y_test_pred=clf.predict(X_test)
```

```python
print(accuracy_score(y_test, y_test_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
train_precision = precision_score(y_train, y_train_pred)
test_precision = precision_score(y_test, y_test_pred)
print("Training Precision:", train_precision)
print("Test Precision:", test_precision)
```

```python
train_f1 = f1_score(y_train, y_train_pred)
test_f1 = f1_score(y_test, y_test_pred)
print("Training F1 Score:", train_f1)
print("Test F1 Score:", test_f1)
```

```python
train_recall = recall_score(y_train, y_train_pred)
test_recall = recall_score(y_test, y_test_pred)
print("Training Recall:", train_recall)
print("Test Recall:", test_recall)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cm, annot=True, fmt="d", cbar=False)
plt.title('SVC Confusion Matrix')
plt.savefig('svc_con_mat')
plt.show()
```
